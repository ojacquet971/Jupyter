```python
def f(x):
  y=x**2-3*x-4
  return y
```

```python
f(-3)
 
```

```python
import numpy as np
import matplotlib. pyplot as plt
x = np.arange(-3,6.5,0.5)
y = f(x)
plt.scatter(x, y, marker='o', color = 'blue')
plt.xlabel ("Abscisses")
plt. ylabel ("Ordonnées")
plt.title("Courbe représentative de f")
x = np. linspace( min(x), max(x), 50)
y2 = f(x)
plt.plot(x, y2, color='green', ls="--")
```
